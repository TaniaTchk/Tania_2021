```python
temperature=[]
for index in range(len(data_lon)):
    param = {'lat':data_lat[index] , 
         'lon':data_lon[index], 
         "exclude": "current,minutely,hourly",
        "units": "metric",
        "appid": '93e1b4d146758830ad3bc42c9a16aeed'}
    
    r= requests.get('https://api.openweathermap.org/data/2.5/onecall', param)
    
    temp=[]
    for i in range (0,7):
        r3=r.json()['daily'][i]['temp']['day']
        temp.append(r3)

    temperature.append(np.mean(temp))

```
